```python
import matplotlib.pyplot as plt
x = range(1,11,1)
cxMarks= [8,9,6,3,5,7,4,1,2,6]
cyMarks= [5,8,9,6,3,2,4,8,8,9]
plt.plot(range(1,11),cxMarks,c="r",marker="o",ls="--" ,label="CX")
plt.plot(range(1,11),cyMarks,c="b",marker=">",ls="--" ,label="CY")
plt.annotate("Lowest",xy=[6.8,1])
plt.annotate("Highest",xy=[8.5,9])
plt.legend()
plt.show()
```

### 04) WAP to demonstrate the use of Pie chart.

```python
import matplotlib.pyplot as plt
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [30, 25, 35, 10]  # Percentage distribution
colors = ['red', 'yellow', 'pink', 'brown']  # Colors for each segment
explode = (0.1, 0, 0, 0)  
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, explode=explode)
plt.title("Fruit Distribution Pie Chart")
plt.show()
```

### 05) WAP to demonstrate the use of Bar chart.

```python
import matplotlib.pyplot as plt
categories = ['Math', 'Science', 'English', 'History', 'Art']
values = [85, 90, 78, 88, 76]  # Example scores
plt.figure(figsize=(8,5))  # Set figure size
plt.bar(categories, values, color=['blue', 'green', 'red', 'purple', 'orange'])

plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Student's Marks in Different Subjects")
plt.show()
```

### 07) WAP to demonstrate the use of Histogram.

```python
import random
SPI = [random.randint(0,10) for _ in range(100)]
plt.hist(SPI,bins=10,histtype='stepfilled',align='mid',label='RESULT')
plt.legend()
plt.show()
```

### 08) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
mon = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
visiters = [11000, 12000, 15000, 12500, 25000, 17000]
bars = plt.bar(mon, visiters, width=0.5)
for i in bars:
    yc = i.get_height()
    plt.text(i.get_x(), yc+100, f"{yc}")
plt.show()
```

### 09) WAP create a Scatter Plot with several colors in Matplotlib?

```python
random.seed(10)
x = [random.random() for i in range(100)]
y = [random.random() for i in range(100)]
z = [random.random() for i in range(100)]
plt.scatter(x,y, c=z,cmap="plasma")
plt.colorbar()
plt.show()
```

### 10) WAP to create a Box Plot.

```python
random.seed(10)
time = [ random.randint(20,120) for i in range(100)]
plt.boxplot(time, vert=False)
plt.show()
```
